# Binomial distribution

so this distribution works on succes and failure 

Suppose a biased coin comes up heads with probability 0.3 when tossed. The probability of seeing exactly x  heads in 6 tosses is

```python
from scipy import stats
x=np.arange(0,7)
x
```

```python
n=6
p=0.3
stats.binom.pmf(x,n,p)
```

```python
plt.bar(x,stats.binom.pmf(x,n,p),color='skyblue',label='PMF (Binomial Distribution)')
plt.xlabel('Probability for outcome')
plt.ylabel('Probability Density')
plt.grid(axis='y',linestyle='--')
plt.tight_layout()
plt.legend(loc='upper right')
plt.show()
```

Calculating atleast or atmost probability of getting heads out of 10

```python
p1=0.5
p2=0.7
n1=20
n2=40
```

```python
x=np.arange(0,41)
x
```

```python
pmf1=stats.binom.pmf(x,n1,p1)
pmf2=stats.binom.pmf(x,n1,p2)
pmf3=stats.binom.pmf(x,n2,p1)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,pmf1,c='blue',marker='^',label=f'p={p1} and n={n1}')
plt.scatter(x,pmf2,c='green',marker='s',label=f'p={p2} and n={n1}')
plt.scatter(x,pmf3,c='red',label=f'p={p1} and n={n2}')
plt.xlabel('Probability for outcome')
plt.ylabel('Probability Density')
plt.grid(axis='y',linestyle='--')
plt.legend(loc='upper right')
plt.show()
```
