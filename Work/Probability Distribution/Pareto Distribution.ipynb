# Pareto Distribution

The Pareto distribution is a power-law probability distribution that is used to describe many types of observable phenomena where a small number of causes or inputs account for a large portion of the effects or outputs. It’s named after Italian economist Vilfredo Pareto, who observed that 80% of Italy’s land was owned by 20% of the population — leading to the well-known 80/20 rule.

## 🎯 Goal:
We want to find the threshold x such that the top 20% of the population owns 80% of the wealth (or other resource).

```python
x_80 = stats.pareto.ppf(0.8, b=alpha, scale=x_m)
x = np.linspace(x_m, 10, 1000)
cdf = stats.pareto.cdf(x, b=alpha, scale=x_m)

plt.figure(figsize=(8, 5))
plt.plot(x, cdf, label='Pareto CDF')
plt.axhline(0.8, color='red', linestyle='--')
plt.axvline(x_80, color='green', linestyle='--', label=f'x at 80% = {x_80:.2f}')
plt.title("80-20 Rule on Pareto Distribution")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_80 = stats.pareto.ppf(0.8, b=alpha, scale=x_m)
x = np.linspace(x_m, 10, 1000)
pdf = stats.pareto.pdf(x, b=alpha, scale=x_m)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Pareto PDF')
plt.axhline(0.8, color='red', linestyle='--')
plt.axvline(x_80, color='green', linestyle='--', label=f'x at 80% = {x_80:.2f}')
plt.title("80-20 Rule on Pareto Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

### 2.You generate 10,000 random samples from a Pareto distribution with parameters alpha = 2.5 and x_m = 1.
How would you empirically verify the 80-20 rule using these samples?

```python
x_m=1
x=np.linspace(x_m,10,1000)
pdf_1000=stats.pareto.pdf(x,2.5,1)
x_80=stats.pareto.ppf(0.8,2.5,1)
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_1000, label='Pareto PDF')
plt.axhline(0.8, color='red', linestyle='--')
plt.axvline(x_80, color='green', linestyle='--', label=f'x at 80% = {x_80:.2f}')
plt.title("80-20 Rule on Pareto Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
