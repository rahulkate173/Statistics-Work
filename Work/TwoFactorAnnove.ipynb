```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr

# --- Generate Example Data (for Demonstration) ---

# Example 1: Positive Monotonic Trend
x1 = np.linspace(0, 1, 100)  # Generate 100 evenly spaced points between 0 and 1
y1 = x1**2 + np.random.normal(0, 0.2, 100)  # Add some noise for a non-perfect trend

# Example 2: Negative Monotonic Trend
x2 = np.linspace(0, 1, 100)
y2 = 1 - x2**1.5 + np.random.normal(0, 0.2, 100)  # Non-linear decreasing trend

# --- Calculate Spearman's Rank Correlation ---

rho1, p_value1 = spearmanr(x1, y1)
rho2, p_value2 = spearmanr(x2, y2)

# --- Plotting ---

plt.figure(figsize=(10, 4))  # Adjust figure size for side-by-side plots

# Plot 1: Positive Correlation
plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
plt.scatter(x1, y1, color='orange', s=30)
plt.title(f"Spearman correlation={rho1:.2f}")  # Display the correlation
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--', alpha=0.5)

# Plot 2: Negative Correlation
plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
plt.scatter(x2, y2, color='orange', s=30)
plt.title(f"Spearman correlation={rho2:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
