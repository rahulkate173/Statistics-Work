 # Q-Q Plot
 ## So this and method to check whether two distribution are same or not where X , Y are random variable for distribution
 ## Mostly use to determine that whether the given distribution id normal or not
 ### How it works ?
 ### => In QQ plot to determine whether it is normal or not we first assume that the second distribution is normal distribution .
 ### 1. The X is the data distribution for which we want to calculate We sort the data and calculated the quantile for each point (ex: 1st quantile ... 100th quantile)
 ### 2. [X,Y] are plotted in scatter plot
 ### 3. If point passes through diagonal then it is normal distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import norm
import statsmodels.api as sm
plt.style.use("ggplot")
color_pal = sns.color_palette()
```

```python
x = np.linspace(-3,3,1000)
```

```python
samples = np.random.normal(0,1,1000)
samples.shape
```

```python
y = norm.pdf(x,0,1) # N(0,1)
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,6))

# PDFs
ax[0].plot(x, y, 'b-', linewidth=2, label='N(0,1)')
# qqplot
sm.qqplot(samples, line='45',ax=ax[1])
# Sample histograms (normalized)
ax[0].hist(samples, bins=30, density=True, alpha=0.5, color=color_pal[3], label='samples')
ax[0].set_xlabel('Samples')
ax[0].set_ylabel('Density')
ax[0].set_title('Probablilty density function ')
ax[0].legend()
ax[0].grid(True, alpha=0.3)
ax[1].set_title("Q-Q plot")
plt.show()
```

```python
## Unfirom distribution
pdf_unifrom = scipy.stats.uniform.pdf(x)
samples = np.random.uniform(0, 1, 1000)
```

```python
plt.figure(figsize=(10, 6))

# PDFs
plt.plot(x, pdf_unifrom, 'b-', linewidth=2, label='N(0,1)')
sm.qqplot(samples, dist=stats.uniform, line='45')
# Sample histograms (normalized)
plt.hist(samples, bins=10, density=True, alpha=0.5, color=color_pal[3], label='samples')
plt.xlabel('Samples')
plt.ylabel('Density')
plt.title('Probablilty density function ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,6))

# PDFs
ax[0].plot(x, pdf_unifrom, 'b-', linewidth=2, label='Uniform : Original')
# qqplot
sm.qqplot(samples, dist=stats.uniform, line='45',ax=ax[1])
# Sample histograms (normalized)
ax[0].hist(samples, bins=10, density=True, alpha=0.5, color=color_pal[3], label='samples')
ax[0].set_xlabel('Samples')
ax[0].set_ylabel('Density')
ax[0].set_title('Probablilty density function ')
ax[0].legend()
ax[0].grid(True, alpha=0.3)
ax[1].set_title("Q-Q plot :  Uniform Distribution")
plt.show()
```
