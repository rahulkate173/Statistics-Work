# Unifrom Distribution
## It is an type of probability distribution where all values are equally likely in given range
## example : If u randomly select value from this distribution then the value is equally likely to other
## example : Dice roll (discreate random variable)
## example : Human height between (75,85) cm ( continuous random variable)
# X ~ U (a,b)
## a = lower bound , b = upper bound


image.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import norm
import statsmodels.api as sm
plt.style.use("ggplot")
color_pal = sns.color_palette()
```

```python
x = np.linspace(-3,3,1000)
samples = np.random.uniform(1,2,1000) # a = 1 , b = 2
samples.shape , x.shape
```

```python
## Unfirom distribution pdf
pdf_unifrom = stats.uniform.pdf(x,loc=1,scale=1)
cdf_uniform = stats.uniform.cdf(x,loc=1,scale=1)
```

```python
pdf_unifrom[np.argmax(pdf_unifrom)]
```

```python
## confirmaing that it is uniform
fig,ax = plt.subplots(1,2,figsize=(10,6))

# PDFs
ax[0].plot(x, pdf_unifrom, 'b-', linewidth=2, label='Uniform : Original')
# qqplot
sm.qqplot(samples, dist=stats.uniform, line='45', ax=ax[1], loc=1, scale=1)
# Sample histograms (normalized)
ax[0].hist(samples, bins=10, density=True, alpha=0.5, color=color_pal[3], label='samples')
ax[0].axhline(pdf_unifrom[np.argmax(pdf_unifrom)], xmin=0, xmax=1, color='r', linestyle='--', linewidth=2) # the prob funtion
ax[0].set_xlabel('Samples')
ax[0].set_ylabel('Density')
ax[0].set_title('Probablilty density function ')
ax[0].legend()
ax[0].grid(True, alpha=0.3)
ax[1].set_title("Q-Q plot :  Uniform Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# PDFs
plt.plot(x, pdf_unifrom, 'b-', linewidth=2, label='Unifrom')
sm.qqplot(samples, dist=stats.uniform, line='45', ax=ax[1], loc=1, scale=1)
# Sample histograms (normalized)
plt.hist(samples, bins=10, density=True, alpha=0.5, color=color_pal[3], label='samples')
plt.xlabel('Samples')
plt.ylabel('Density')
plt.title('Probablilty density function ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# PDFs
plt.plot(x, cdf_uniform, 'b-', linewidth=2, label='Unifrom')
# Sample histograms (normalized)
plt.hist(samples, bins=10, density=True, alpha=0.5, color=color_pal[3], label='samples')
plt.xlabel('Samples')
plt.ylabel('Density')
plt.title('Cummalitative density function ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# PDFs
plt.plot(x, pdf_unifrom, 'b-', linewidth=2, label='Unifrom-pdf')
plt.plot(x, cdf_uniform, 'g--', linewidth=2, label='Unifrom-cdf')
# Sample histograms (normalized)
plt.hist(samples, bins=10, density=True, alpha=0.5, color=color_pal[3], label='samples')
plt.xlabel('Samples')
plt.ylabel('Density')
plt.title('PDF AND CDF ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## use cases
1. Hyperparameter Tuning (Random Search)
When training complex models like neural networks, you need to find the best configuration of parameters (hyperparameters), such as the learning rate or batch size.
Use Case: In a random search algorithm, you define a range of possible values for each hyperparameter. To ensure an unbiased search across the entire space, these values are sampled from a uniform distribution, giving every potential value within the specified range an equal chance of being tested. This method is often more efficient than a rigid grid search.
2. Weight Initialization in Neural Networks
The starting weights of a neural network must be initialized with random values to break symmetry between neurons and allow the network to learn effectively.
Use Case: Weights are frequently initialized using a uniform distribution within a carefully selected small range (e.g., in the Xavier or Kaiming initialization methods). This helps prevent vanishing or exploding gradient problems during the initial phases of training by ensuring the variance of activations remains stable across layers.
3. Data Simulation and Random Sampling The uniform distribution is the foundation of most computer-generated randomness. Use Case: In simulations (like Monte Carlo methods), a standard uniform distribution \(U(0,1)\) is used to generate core random numbers. These uniform random numbers can then be mathematically transformed using the inverse transform sampling method to generate data following any other complex distribution (e.g., normal, exponential, etc.) that the machine learning model might require for training or testing.
