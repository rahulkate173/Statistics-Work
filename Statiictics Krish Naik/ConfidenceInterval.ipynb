```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stat
```

# Confidence Interval

- Point Estimate: The value of any statistics that estimate value of parameter

sigma=100,alpha=0.05,n=25,x_bar=520

```python
sigma=100
alpha=0.05
n=30
x_bar=520
```

point estimate + margin error

1. Population Standard deviation is given
2. n>=30

- upper bound = x_bar + z_alpha/2 * (sigma/sqrt_n)
- lower bound = x_bar - z_alpha/2 * (sigma/sqrt_n)

```python
z_critical=stat.norm.ppf(1-alpha/2)
margin_error=z_critical*(sigma/np.sqrt(n))
ub=x_bar+margin_error
lb=x_bar-margin_error
x =np.linspace(x_bar-4*sigma,x_bar+4*sigma)
y = stat.norm.pdf(x,x_bar,sigma)
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Normal Distribution", color="blue")
plt.axvline(x_bar, color="red", linestyle="dashed", label="Mean (Î¼)")

plt.axvline(lb, color="green", linestyle="dashed", label="Lower CI Bound")
plt.axvline(ub, color="green", linestyle="dashed", label="Upper CI Bound")
plt.title('Normal Distribution')
plt.grid(True)
plt.show()
```

```python
# priotize in function
def range_calculate(x_bar,sigma,n,alpha):
    z_critical=stat.norm.ppf(1-alpha/2)
    margin_error=z_critical*(sigma/np.sqrt(n))
    ub=x_bar+margin_error
    lb=x_bar-margin_error
    return ub,lb
def distribution(ub,lb,alpha,sigma,x_bar):
    x =np.linspace(x_bar-4*sigma,x_bar+4*sigma)
    y = stat.norm.pdf(x,x_bar,sigma)
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Normal Distribution", color="blue")
    plt.axvline(x_bar, color="red", linestyle="dashed", label="Mean (Î¼)")
    plt.axvline(lb, color="green", linestyle="dashed", label="Lower CI Bound")
    plt.axvline(ub, color="green", linestyle="dashed", label="Upper CI Bound")
    plt.title('Normal Distribution')
    plt.grid(True)
    plt.show()
```

```python
ub,lb=range_calculate(x_bar,sigma,n,alpha)
distribution(ub,lb,alpha,sigma,x_bar)
```
